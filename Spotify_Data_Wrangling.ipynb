```python
#The first dataset (data1) is the main dataset everything will be based off of, so I'll check to make sure all of the songs
#from the other datasets are in Data 1
titles2020 = data2['track_name']
titles2021 = data3['track_name']
titles2010s = data4['title']
```

```python
df_mod = data1
```

```python
df_mod['check_true'] = df_mod['song_name'].isin(titles2020)
```

```python
df_mod = df_mod[df_mod['check_true'] == True] 
```

```python
df_mod
```

```python
#It's a good thing to be noticed early on, but it looks like the main DataFrame has a LOT of duplicates across various
#areas of the dataset, so I'm going to deal with those before I get any further

#I could also try to get things from the Spotify API, but I as the Spotify API does not have a popularity category it would
#be impossible to move forward with the project, so I will just do my best to get as much of the info in as I can
data_mod = data1[data1.duplicated(subset = ['song_name']) != True]
#What songs were duplicated?
data_dup = data1[data1.duplicated(subset = ['song_name']) == True]
```

```python
#Some were even duplicated mutiple times. It's a good thing to check and get rid of this
data_dup[data_dup.duplicated() == True]
```

```python
#Seems like there are quite a few things missing. I will try to match things based on similarity, but if that doesn't work
#then this is how it is
data_mod[data_mod['check_true'] == True]
```

```python
#Making one list for all song names in all three top songs lists
pop_song_names = []
for index, item in titles2020.items():
    pop_song_names.append(item)
for index, item in titles2021.items():
    pop_song_names.append(item)
for index, item in titles2010s.items():
    pop_song_names.append(item)
print(len(pop_song_names))
print(pop_song_names[:50])
```

```python
#703 of the songs seems to be a fair number, considering the 2010's list was not exclusivly top 50 but just top songs
#Additionally, as some of the values have questionable titles in place and won't relate to the main list, it should be fine
for item in pop_song_names:
    print(process.extract(item, data_mod['song_name'], limit = len(data_mod['song_name']))[1])
#This method took too long and did not have results that were ideal, so I will just scrape what are exact matches. 
#The reason I think there is such large and random gaps in the data is due to how the people who put it online scraped 
#stuff from the Spotify API, as you generally have to specify the song you want to get. It would also explain the repition
```

```python
#For now I will just take the songs that I know are for sure in the list, as there doesn't seem to be much to gain from 
#looking at the closeness 
data_mod['check_true'] = data_mod['song_name'].isin(pop_song_names)
data_mod[data_mod['check_true'] == True]
```

```python
#It looks like many of the titles were not included in this dataset, but that is to be expected from a random grab of songs
#With this, we have a functional row to show if a song made it to the top 50 or not. Lets look at the info of the new df 
#and give it a more proper and less redundant name
df_main = data_mod
df_main.rename(columns = {'check_true':'Is_top_track'}, inplace = True)
print(df_main.info())
```

```python
#Ensure no repeats
print(len(df_main['song_name'].unique()))
```

```python
#Thankfully there doesn't seem to be any holes in the data and all of the values seem to be correct for the given category
#So I'm going to ensure that the vals are all within the scope set up by the inital dataset, meaning that
#Song name is a string (this is already true, so we can ignore this)
#Song popularity is an int between 0 and 100
#Song duration is and in between 0 and 1800000 (not actually enforced but 99.9% of the songs should be less than 30 mins long)
#Acousticness as a float on a scale from 0-1
#Danceability as a float on a scale from 0-1
#Energy as a float on a scale from 0-1
#Instrumentalness as a float on a scale from 0-1
#Key as a int on a scale from 0-11 
#Liveness as a float from 0-1
#Loudness as a float from -38.768  
#Audio Mode as a int either as 0 or 1 
#Speechiness as a float from 0-1   
#Tempo as a float from 1-250(no values exceed 250), 0 vals indicate it is not actually a song, so those could be dropped
#Time signature as int at 1-5 though most will be 4
#Audio valence as a float from 0-1

#For now, lets just describe the whole DF to get an Idea of what to do from there
df_main.describe()
```

```python
#Song popularity looks fine, but there is a really long song looking at the duration, lets see what that is and if we should
#keep it

#Looking at the longest song, it is real and it is really this long(at least the spotify version), not my style, but has
#some good parts. This leads me to belive that song duration category is okay.
df_main[df_main['song_duration_ms'] == df_main['song_duration_ms'].max()]
```

```python
#The min category looks pretty odd, most so for the tempo category like there is something that is not a song in the dataset
#Looking at it
df_main[df_main['tempo'] == df_main['tempo'].min()]
```

```python
#One of those is definitly not a song, and the other seems to have some problems in the entry (having no valence doesn't 
#really happen) Ill be removing these and then looking at the stats again
df_main.drop([7119, 11171], axis=0, inplace=True)
df_main.describe()
```

```python
#While it may seem slow, that is to be expected of the minimum speed song
df_main[df_main['tempo'] == df_main['tempo'].min()]
```
